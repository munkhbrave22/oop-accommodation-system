```python
#Base Class
class Accommodation:

  #implement a class variable
  accommodation_number = 0

  def __init__(self, n, p, r, l):
    self.name = n
    self.price = p
    self.rating = r
    self.location = l

    Accommodation.accommodation_number += 1

  #Getters and Setters
  #Getters
  def get_name(self):
    return self.name

  def get_price(self):
    return self.price

  def get_rating(self):
    return self.rating

  def get_location(self):
    return self.location

  #Setters
  def set_price(self, new_price):
    if new_price > 0:
      self.price = new_price

  def set_rating(self, new_rating):
    if new_rating <= 5:
      self.rating = new_rating

  #Discount setter
  def apply_discount(self, percentage):
    if 0 < percentage < 100:
      self.price *= (1 - percentage / 100)
```

```python
class Hotel(Accommodation):
  def __init__(self, name, price_per_night, rating, location, stars):
    super().__init__(name, price_per_night, rating, location)
    self.__stars = stars

  def get_stars(self):
    return self.__stars
```

```python
#Subclass 2
class Apartment(Accommodation):
  def __init__(self, name, price_per_night, rating, location, rooms):
    super().__init__(name, price_per_night, rating, location)
    self.__rooms = rooms

  def get_rooms(self):
    return self.__rooms
```

```python
#Object Creation & Method Calls
# Base class object
acc = Accommodation("City Stay", 100, 4.2, "Berlin")

# Subclass objects
hotel = Hotel("Grand Hotel", 180, 4.6, "Paris", 5)
apartment = Apartment("Sunny Flat", 90, 4.3, "Madrid", 3)

# Method calls
acc.apply_discount(10)
hotel.set_rating(4.8)

print(acc.get_price())
print(hotel.get_stars())
print(apartment.get_rooms())

print("Total accommodations:", Accommodation.accommodation_number)
```
